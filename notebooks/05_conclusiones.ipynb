# Notebook 05 — Conclusiones

**Objetivo:** Responder las preguntas definidas al inicio del proyecto con evidencia obtenida durante el análisis, diferenciando evidencia, interpretación y conclusión. Las conclusiones deben ser coherentes con el informe final y la aplicación Streamlit.

## Pregunta 1: ¿Fumar es el factor que más influye en los costos médicos?

**Evidencia obtenida:**
- El costo medio de pacientes fumadores (~32.000 USD) es casi 4 veces mayor al de no fumadores (~8.400 USD).
- La correlación de `smoker_num` con `charges` es r≈0.79, la más alta de todas las variables.
- En el análisis multivariado, el patrón se mantiene en todas las regiones geográficas.
- En el PCA, PC1 (el componente de mayor varianza explicada) está dominada por `smoker_num` y `charges`.

**Interpretación:** El tabaquismo no solo eleva los costos de forma directa, sino que constituye el eje de variación más importante en el dataset, visible tanto en el análisis bivariado como en la reducción de dimensionalidad.

**Conclusión:** Sí. El tabaquismo es el factor con mayor influencia sobre los costos médicos en este dataset. La diferencia no es marginal: los fumadores generan costos casi cuatro veces superiores independientemente de la región, el sexo o la cantidad de hijos.

## Pregunta 2: ¿A mayor IMC y edad, mayores costos? ¿Se potencian entre sí?

**Evidencia obtenida:**
- La correlación de `age` con `charges` es r≈0.30; la de `bmi` es r≈0.20. Ambas positivas pero moderadas.
- Los scatterplots muestran que dentro del grupo de fumadores, los costos más altos corresponden a pacientes de mayor edad y mayor IMC.
- En el PCA, PC2 captura principalmente la variación de `age` y `bmi`, independiente del efecto del tabaquismo (PC1).

**Interpretación:** La edad y el IMC tienen una relación positiva con los costos, pero de magnitud moderada. Su efecto se potencia principalmente dentro del grupo de fumadores: un fumador mayor con IMC elevado representa el perfil de mayor costo en el dataset.

**Conclusión:** A mayor edad y mayor IMC, los costos tienden a ser mayores. Sin embargo, el efecto es moderado y secundario respecto al tabaquismo. La potenciación entre ambas variables ocurre principalmente en pacientes fumadores.

## Pregunta 3: ¿Existen grupos diferenciados de pacientes según su perfil de riesgo?

**Evidencia obtenida:**
- La distribución de `charges` presenta asimetría positiva con una estructura bimodal, lo que sugiere al menos dos grupos con perfiles distintos.
- La proyección PCA en PC1-PC2 muestra una separación espacial clara entre fumadores y no fumadores.
- El gradiente de costos está alineado con PC1, confirmando que la separación refleja diferencias reales en el nivel de costos.

**Interpretación:** El dataset contiene al menos dos grupos estructuralmente diferenciados: pacientes de bajo riesgo (no fumadores, costos bajos) y pacientes de alto riesgo (fumadores, costos elevados). Dentro de cada grupo, edad e IMC introducen variación adicional.

**Conclusión:** Sí existen grupos diferenciados. El eje principal de separación es el tabaquismo. Esta estructura es identificable tanto en el análisis descriptivo como en el espacio de componentes principales.

## Limitaciones

- El alcance de las conclusiones se encuentra condicionado por la información disponible en el dataset y por las decisiones documentadas durante el proceso. No se dispone de información clínica adicional (diagnósticos, historial médico) que permitiría ampliar el análisis.
- La imputación por mediana en `age` y `bmi` introduce una simplificación que podría subestimar la variabilidad real en esas variables.
- El análisis es de carácter exploratorio: las relaciones observadas no permiten establecer causalidad, únicamente asociación estadística.
- El dataset no incluye información temporal, lo que impide analizar evolución de costos en el tiempo.

## Mejoras futuras

- Una mejora futura podría consistir en incorporar variables clínicas adicionales que permitan ampliar el alcance del análisis de perfiles de riesgo.
- La aplicación de técnicas de clustering (k-means, DBSCAN) sobre el espacio PCA permitiría formalizar la segmentación de pacientes observada.
- Podría explorarse la construcción de modelos predictivos de costos una vez consolidada la comprensión del dataset.
